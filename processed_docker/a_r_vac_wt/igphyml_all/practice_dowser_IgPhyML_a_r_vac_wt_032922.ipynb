```r
# IgPhyML visualization in R
# graphing individual data

#!/usr/bin/R
library(alakazam)
library(ggplot2)
library(ape)
library(dplyr)
library(igraph)
library(dowser)

#read in three different samples
db_a = readIgphyml("a_heavy_germ-pass_igphyml-pass.tab",id="a")
db_r = readIgphyml("r_heavy_germ-pass_igphyml-pass.tab",id="r")
db_vac = readIgphyml("vac_heavy_germ-pass_igphyml-pass.tab",id="vac")
db_wt = readIgphyml("wt_heavy_germ-pass_igphyml-pass.tab",id="wt") 

#combine objects into a dataframe
# comb = combineIgphyml(list(db_a, db_r, db_vac),format="wide")
```

```r
# Build maxmimum parsimony trees for first four clones using 
# phangorn package in R
trees <- plot(db_a$trees[[1]], layout=layout_as_tree, edge.label=E(db_a$trees[[1]])$weight)
```

```r
# Read in and plot a tree from an IgPhyML
# db_r
print(db_r$param$OMEGA_CDR_MLE[1])
plot(db_r$trees[[1]], layout=layout_as_tree, edge.label=E(db_r$trees[[1]])$weight)
```

```r
# Read in and plot a tree from an IgPhyML
# db_vac
print(db_vac$param$OMEGA_CDR_MLE[1])
plot(db_vac$trees[[1]], layout=layout_as_tree, edge.label=E(db_vac$trees[[1]])$weight)
```

```r
# Read in and plot a tree from an IgPhyML
# db_wt
print(db_wt$param$OMEGA_CDR_MLE[1])
plot(db_wt$trees[[1]], layout=layout_as_tree, edge.label=E(db_wt$trees[[1]])$weight)
```

```r
# manual color codes
sample_colors <- c("Unprotected"="deeppink3", "Vaccinated"="steelblue2", "Control"="slategrey")

df <- read.csv("heavy_chain.csv")
head(df)

df <- melt(df ,  id.vars = "isotype", variable.name = 'series')

#create line plot for each column in data frame
ggplot(df, aes(isotype, value)) + theme_bw() + 
        geom_point(aes(colour = series, size=4)) + 
        scale_color_manual(values=sample_colors) + 
        theme(axis.text.x=element_text(angle=45, hjust=1, vjust=1)) +
        ylab("Isotype frequency") + xlab("Paired chains") +
        ggtitle("Paired isotype") +
        theme(legend.position="none")
```
